```python
import pandas as pd
import numpy as np

df = pd.read_csv("reports_sales_listings_item (7).csv")

df["Order Qty"] = np.where(
    (df["Stock"] <= 0) & (df["Sold"] == 1),
    2,
    np.where(
        df["Stock"] <= 0,
        df["Sold"] * (3/4),
        np.where(
            df["Stock"] < 2,
            df["Sold"] / 2,
            np.where(
                (df["Sold"] / 2) - df["Stock"] < 0,
                0,
                (df["Sold"] / 2) - df["Stock"]
            )
        )
    )
)

df["Order Qty"] = np.maximum(0, np.ceil(df["Order Qty"])).astype(int)

df = df[df["Order Qty"] > 0]

df = df.sort_values("Order Qty", ascending=False)


df.to_csv("output.csv", index=False)
```
