```python
import pandas as pd
import numpy as np

# Read the CSV
df = pd.read_csv("input.csv")

# Calculate Order Qty using your report formula
df["Order Qty"] = np.where(
    (df["Stock"] <= 0) & (df["Sold"] == 1),
    2,
    np.where(
        df["Stock"] <= 0,
        df["Sold"] * (3/4),
        np.where(
            df["Stock"] < 2,
            df["Sold"] / 2,
            np.where(
                (df["Sold"] / 2) - df["Stock"] < 0,
                0,
                (df["Sold"] / 2) - df["Stock"]
            )
        )
    )
)

# Round up to the nearest whole number and prevent negatives
df["Order Qty"] = np.maximum(0, np.ceil(df["Order Qty"])).astype(int)

# Optional: Keep only items that need to be ordered
df = df[df["Order Qty"] > 0]

# Sort by largest order quantity first
df = df.sort_values("Order Qty", ascending=False)

# Save the results
df.to_csv("output.csv", index=False)

print(f"Created output.csv with {len(df)} items to order.")
```
